```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense

model = keras.Sequential([
                          Dense(1, activation="relu"),
                          Dense(1, activation="relu")
                          ])  

x = model(tf.ones((1, 1)))

model.summary()
```
